```python
class Solution:
    def totalNQueens(self, n: int) -> int:
        self.n = n
        self.result = 0
        possible_moves = {i : {j for j in range(n)} for i in range(n)}
        for i in range(n//2):
            self.make_a_move(0, i, copy.deepcopy(possible_moves), n)
        self.result *= 2
        if n % 2 == 1:
            self.make_a_move(0, n//2, copy.deepcopy(possible_moves), n)
        return self.result

    def make_a_move(self, i, j, possible_moves, iterations):
        if iterations == 1:
            self.result += 1
            return
        for k in range(self.n):
            possible_moves[k].discard(j)
        possible_moves[i] = set()
        for k in range(i+j+1):
            if k < self.n:
                possible_moves[k].discard(i+j-k)
        if i >= j:
            for k in range(self.n-i+j):
                if i-j+k in possible_moves:
                    possible_moves[i-j+k].discard(k)
        else:
            for k in range(self.n+i-j):
                if k in possible_moves:
                    possible_moves[k].discard(-i+j+k)
        move = self.minimum_moves(copy.deepcopy(possible_moves))
        if move:
            for j in possible_moves[move]:
                self.make_a_move(move, j, copy.deepcopy(possible_moves), iterations-1)
        
    def minimum_moves(self, possible_moves):
        move = (None, 9)
        for key in possible_moves:
            if 0 < len(possible_moves[key]) < move[1]:
                move = (key, len(possible_moves[key]))
        return move[0]
```

```python
# Not mine, but much better answer. More simple and efficient.
class Solution:
    def totalNQueens(self, n: int) -> int:
        self.count = 0
        queens = set()
        diagonals = set()
        anti_diagonals = set()
        self.solveNQueensHelper(n, 0, queens, diagonals, anti_diagonals)
        return self.count

    def solveNQueensHelper(self, n, row, queens, diagonals, anti_diagonals):
        if row == n:
            self.count += 1
            return

        for col in range(n):
            if col in queens or (row - col) in diagonals or (row + col) in anti_diagonals:
                continue

            queens.add(col)
            diagonals.add(row - col)
            anti_diagonals.add(row + col)

            self.solveNQueensHelper(n, row + 1, queens, diagonals, anti_diagonals)

            queens.remove(col)
            diagonals.remove(row - col)
            anti_diagonals.remove(row + col)
```
