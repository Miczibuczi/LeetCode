```python
class Solution:
    def titleToNumber(self, columnTitle: str) -> int:
        separate_numbers = []
        for i in columnTitle:
            if i == "A":
                separate_numbers.append(1)
            elif i == "B":
                separate_numbers.append(2)
            elif i == "C":
                separate_numbers.append(3)
            elif i == "D":
                separate_numbers.append(4)
            elif i == "E":
                separate_numbers.append(5)
            elif i == "F":
                separate_numbers.append(6)
            elif i == "G":
                separate_numbers.append(7)
            elif i == "H":
                separate_numbers.append(8)
            elif i == "I":
                separate_numbers.append(9)
            elif i == "J":
                separate_numbers.append(10)
            elif i == "K":
                separate_numbers.append(11)
            elif i == "L":
                separate_numbers.append(12)
            elif i == "M":
                separate_numbers.append(13)
            elif i == "N":
                separate_numbers.append(14)
            elif i == "O":
                separate_numbers.append(15)
            elif i == "P":
                separate_numbers.append(16)
            elif i == "Q":
                separate_numbers.append(17)
            elif i == "R":
                separate_numbers.append(18)
            elif i == "S":
                separate_numbers.append(19)
            elif i == "T":
                separate_numbers.append(20)
            elif i == "U":
                separate_numbers.append(21)
            elif i == "V":
                separate_numbers.append(22)
            elif i == "W":
                separate_numbers.append(23)
            elif i == "X":
                separate_numbers.append(24)
            elif i == "Y":
                separate_numbers.append(25)
            elif i == "Z":
                separate_numbers.append(26)
        
        separate_numbers = separate_numbers[::-1]
        result = 0
        for j in range(len(separate_numbers)):
            result += 26**j*separate_numbers[j]
        
        return result
```
