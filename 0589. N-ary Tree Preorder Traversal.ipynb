```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val=None, children=None):
        self.val = val
        self.children = children
"""

class Solution:
    def preorder(self, root: 'Node') -> List[int]:
        self.answer = []
        def Traversal(node):
            if not node:
                return
            self.answer.append(node.val)
            for child in node.children:
                Traversal(child)
        Traversal(root)
        return self.answer
```
