```python
# Very, very complicated, but working solution
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        def helper(cur, prev, money, cooldown):
            if cur == prev:
                self.options.add((prev, money, False))
            elif cur < prev:
                if cooldown:
                    self.options.add((prev, money, False))
                else:
                    self.options.add((cur, money, False))
            elif cur > prev:
                self.options.add((cur, money+cur-prev, True))
                self.options.add((prev, money, False))
        
        def pick_best(options):
            best_True = False
            best_False = False
            for option in options:
                if option[2] == False:
                    if not best_False:
                        best_False = option
                        best_False_score = option[1]-option[0]
                        best_money = option
                    else:
                        False_score = option[1]-option[0]
                        if False_score > best_False_score:
                            best_False = option
                            best_False_score = False_score
                        if option[1] > best_money[1]:
                            best_money = option
                elif option[2] == True:
                    if not best_True:
                        best_True = option
                        best_True_score = option[1]-option[0]
                    else:
                        True_score = option[1]-option[0]
                        if True_score > best_True_score:
                            best_True = option
                            best_True_score = True_score
            self.options.clear()
            self.options.add(best_False)
            self.options.add(best_True)
            self.options.add(best_money)


        self.options = {(prices[0], 0, False)}
        for i in range(1, len(prices)):
            cur_options = self.options.copy()
            self.options.clear()
            for option in cur_options:
                p, m, cd = option
                helper(prices[i], p, m, cd)
            if len(self.options) >= 20:
                pick_best(self.options)
        ans = 0
        for option in self.options:
            ans = max(ans, option[1])
        return ans
```
