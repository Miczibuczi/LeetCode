```python
import pandas as pd

def find_primary_department(employee: pd.DataFrame) -> pd.DataFrame:
    seen = set()
    seen_2 = set()
    for row in employee.itertuples():
        if row.employee_id not in seen:
            seen.add(row.employee_id)
        else:
            seen_2.add(row.employee_id)
    seen = seen - seen_2
    employee = employee[(employee["primary_flag"] == "Y") | (employee["employee_id"].isin(seen))]
    return employee[["employee_id", "department_id"]]
```
