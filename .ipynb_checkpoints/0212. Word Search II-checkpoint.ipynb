```python
class Tree:
    def __init__(self, value=None, children=None):
        self.val = value
        self.children = children if children is not None else set()

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        def add_letter(cur, i, j, seen):
            cur_seen = seen.copy()
            cur_seen.add((i, j))
            for child in cur.children:
                if child.val == board[i][j]:
                    new_cur = child
                    break
            else:
                new_cur = Tree(board[i][j])
                cur.children.add(new_cur)
            if i > 0 and (i-1, j) not in seen:
                add_letter(new_cur, i-1, j, cur_seen)
            if i+1 < len(board) and (i+1, j) not in seen:
                add_letter(new_cur, i+1, j, cur_seen)
            if j > 0 and (i, j-1) not in seen:
                add_letter(new_cur, i, j-1, cur_seen)
            if j+1 < len(board[0]) and (i, j+1) not in seen:
                add_letter(new_cur, i, j+1, cur_seen)

        root = Tree()
        for i in range(len(board)):
            for j in range(len(board[0])):
                add_letter(root, i, j, set())
        ans = []
        for word in words:
            cur = root
            for letter in word:
                for child in cur.children:
                    if child.val == letter:
                        cur = child
                        break
                else:
                    break
            else:
                ans.append(word)
        return ans
```

```python
class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        my_dict = defaultdict(set)
        for i in range(len(board)):
            for j in range(len(board[0])):
                my_dict[board[i][j]].add((i, j))
        
        ans = []
        for word in words:
            for letter in word:
                if letter not in my_dict:
                    break
            else:
                stack = [item+(set(),) for item in my_dict[word[0]]]
                for letter in word[1:]:
                    new_stack = []
                    for item in stack:
                        i, j, seen = item
                        seen_copy = seen.copy()
                        seen_copy.add((i, j))
                        if i > 0 and board[i-1][j] == letter and (i-1, j) not in seen:
                            new_stack.append((i-1, j, seen_copy))
                        if i+1 < len(board) and board[i+1][j] == letter and (i+1, j) not in seen:
                            new_stack.append((i+1, j, seen_copy))
                        if j > 0 and board[i][j-1] == letter and (i, j-1) not in seen:
                            new_stack.append((i, j-1, seen_copy))
                        if j+1 < len(board[0]) and board[i][j+1] == letter and (i, j+1) not in seen:
                            new_stack.append((i, j+1, seen_copy))
                    stack = new_stack
                if stack:
                    ans.append(word)
        return ans
```
