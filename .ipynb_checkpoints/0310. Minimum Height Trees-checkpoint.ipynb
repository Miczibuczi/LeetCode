```python
# Best answer, 392 ms
class Solution:
    def findMinHeightTrees(self, n: int, edges: List[List[int]]) -> List[int]:
        if not edges:
            return [0]
        pointers = [[] for i in range(n)]
        for edge in edges:
            a, b = edge
            pointers[a].append(b)
            pointers[b].append(a)
        next_loop = set()
        count = n        
        seen = set()
        for i in range(n):
            if len(pointers[i]) == 1:
                seen.add(i)
                next_loop.add(pointers[i][0])
                count -= 1
        if count == 0:
            return seen
        while count > 0:
            end_of_loop = True
            cur_ans = []
            cur_loop = next_loop.copy()
            next_loop.clear()
            for num in cur_loop:
                just_once = False
                for i in pointers[num]:
                    if i not in seen:
                        if just_once:
                            next_loop.add(num)
                            break
                        else:
                            just_once = i+1
                else:
                    cur_ans.append(num)
                    next_loop.add(just_once-1)
                    count -= 1
                    end_of_loop = False
            for num in cur_ans:
                seen.add(num)
            if end_of_loop:
                return cur_loop
        return cur_ans
```

```python
# First answer, slow, 4441 ms
class Solution:
    def findMinHeightTrees(self, n: int, edges: List[List[int]]) -> List[int]:
        if not edges:
            return [0]
        count = dict()
        pointers = [[] for i in range(n)]
        for edge in edges:
            a, b = edge
            if a not in count:
                count[a] = 1
            else:
                count[a] += 1
            if b not in count:
                count[b] = 1
            else:
                count[b] += 1
            pointers[a].append(b)
            pointers[b].append(a)

        while len(set(count.values())) != 1:
            cur_count = count.copy()
            for num, occur in cur_count.items():
                if occur == 1:
                    del count[num]
                    for pointer in pointers[num]:
                        if pointer in count:
                            count[pointer] -= 1
                    pointers[num] = []
        return count.keys()
```
