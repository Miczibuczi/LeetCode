```python
import pandas as pd

def capital_gainloss(stocks: pd.DataFrame) -> pd.DataFrame:
    stocks["price"] = stocks.apply(lambda row: row.price if row.operation=="Sell" else -row.price, axis=1)
    stocks = stocks.groupby(by="stock_name").agg(capital_gain_loss=("price", "sum")).reset_index()
    return stocks[["stock_name", "capital_gain_loss"]]
```
