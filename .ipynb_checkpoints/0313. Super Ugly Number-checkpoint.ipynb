```python
# Best answer, 376ms
class Solution:
    def nthSuperUglyNumber(self, n: int, primes: List[int]) -> int:
        if n == 1:
            return n
        ugly = primes.copy()
        heapify(ugly)
        anty_ugly = [-x for x in primes][::-1]
        heapify(anty_ugly)
        seen = {x for x in primes}
        for i in range(n-2):
            cur = heappop(ugly)
            for prime in primes:
                new_ugly = cur*prime
                if new_ugly in seen:
                    continue
                seen.add(new_ugly)
                heappush(anty_ugly, -new_ugly)
                if len(anty_ugly) > n-1:
                    x = heappop(anty_ugly)
                    if -x == new_ugly:
                        break
                heappush(ugly, new_ugly)
        return ugly[0]
```

```python
# First answer, slow, 2912ms
class Solution:
    def nthSuperUglyNumber(self, n: int, primes: List[int]) -> int:
        if n == 1:
            return n
        ugly = primes.copy()
        seen = {x for x in primes}
        for i in range(n-2):
            cur = ugly.pop(0)
            for prime in primes:
                new_ugly = cur*prime
                if new_ugly in seen:
                    continue
                seen.add(new_ugly)
                index = bisect_right(ugly, new_ugly)
                if index > n-i-2:
                    break
                ugly.insert(index, new_ugly)
        return ugly[0]
```

```python
# Memory limit exceeded
class Solution:
    def nthSuperUglyNumber(self, n: int, primes: List[int]) -> int:
        if n == 1:
            return 1

        def count_num(num_components):
            out = 1
            for i in range(len(primes)):
                out *= primes[i]**num_components[i]
            return out

        cur_nums = [primes[0]]
        my_dict = defaultdict(list)
        seen = set()
        my_dict[primes[0]].append([1]+[0 for i in range(len(primes)-1)])
        for i in range(n-2):
            num = cur_nums.pop(0)
            num_components_list = my_dict[num]
            for num_components in num_components_list:
                new_num_components = num_components.copy()
                new_num_components[0] += 1
                new_num = count_num(new_num_components)
                my_dict[new_num].append(new_num_components)
                if new_num not in seen:
                    seen.add(new_num)
                    insort(cur_nums, new_num)

                for i, component in enumerate(num_components):
                    if component != 0 and i != len(primes)-1:
                        new_num_components = num_components.copy()
                        new_num_components[i] -= 1
                        new_num_components[i+1] += 1
                        new_num = count_num(new_num_components)
                        my_dict[new_num].append(new_num_components)
                        if new_num not in seen:
                            seen.add(new_num)
                            insort(cur_nums, new_num)
        return cur_nums[0]
```
