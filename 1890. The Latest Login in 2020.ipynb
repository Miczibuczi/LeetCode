```python
import pandas as pd

def latest_login(logins: pd.DataFrame) -> pd.DataFrame:
    logins = logins[logins["time_stamp"].between(pd.Timestamp("2020-01-01 00:00:00"), pd.Timestamp("2020-12-31 23:59:59"), inclusive="both")]
    logins = logins.groupby(by="user_id").agg(last_stamp=("time_stamp", "max")).reset_index()
    return logins
```

```python
# More time-efficient solution generated by ChatGPT:
import pandas as pd

def latest_login(logins: pd.DataFrame) -> pd.DataFrame:
    logins = logins[
        (logins["time_stamp"] >= pd.Timestamp("2020-01-01")) & 
        (logins["time_stamp"] <= pd.Timestamp("2020-12-31 23:59:59"))
    ]
    latest_logins = logins.groupby("user_id")["time_stamp"].max().reset_index(name="last_stamp")
    return latest_logins
```
