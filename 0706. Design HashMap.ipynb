```python
class Node:
    def __init__(self, val, key, next = None):
        self.val = val
        self.key = key
        self.next = next

class MyHashMap:

    def __init__(self):
        self.size = 10**5
        self.array = [Node(None, None)]*self.size

    def put(self, key: int, value: int) -> None:
        index = key % self.size
        if self.array[index].key == None:
            self.array[index] = Node(value, key)
        elif self.array[index].key == key:
            self.array[index].val = value
        else:
            head = self.array[index]
            while head is not None:
                if head.next is None:
                    head.next = Node(value, key)
                    break
                elif head.next.key == key:
                    head.next.val = value
                    break
                head = head.next


    def get(self, key: int) -> int:
        index = key % self.size
        head = self.array[index]
        while head is not None:
            if head.key == key:
                return head.val
            head = head.next
        return -1

    def remove(self, key: int) -> None:
        index = key % self.size
        head = self.array[index]
        if not head:
            return
        elif head.key == key:
            self.array[index] = self.array[index].next
            if self.array[index] is None:
                self.array[index] = Node(None, None)
        else:
            while head.next:
                if head.next.key == key:
                    head.next = head.next.next
                    if head.next is None:
                        head.next = Node(None, None)
                    break
                head = head.next


# Your MyHashMap object will be instantiated and called as such:
# obj = MyHashMap()
# obj.put(key,value)
# param_2 = obj.get(key)
# obj.remove(key)
```
